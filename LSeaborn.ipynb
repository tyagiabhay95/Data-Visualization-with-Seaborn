```python
import seaborn as sns
import pandas as pd
```

# Distribution Plot

## *distplot *jointplot *pairplot

```python
df=sns.load_dataset("tips")
#pd.options.display.max_rows = None
df
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

#### Join Plot

```python
#Using jointplot we have hex shape as well as histogram to show data concentration.
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

### Pair Plot

```python
# Pair plot is a 'scatter plot' in which on variable is matched with various other variables.
sns.pairplot(df,hue='sex')
#sns.pairplot(df,hue='smoker')
```

# Categorical Plots

#### boxplot *violinplot *countplot *barplot

```python
#Countplot
sns.countplot('sex',data=df)
#sns.countplot('day',data=df)
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(data=df,orient='h')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
#Violin plot helps us to see both the distribution of data in terms of kernel density estimation and box plot.
sns.violinplot(x='total_bill',y='day',data=df)
```
